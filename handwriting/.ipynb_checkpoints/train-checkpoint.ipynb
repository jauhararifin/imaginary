```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('data.csv').values
data_x = data[:,:-1]
data_y = data[:,-1]
```

```python
clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(100, 120), alpha=1e-5, random_state=1)
clf.fit(data_x, data_y)
prediction = clf.predict(data_x)

print(accuracy_score(data_y, prediction))
print(prediction)
```
